```python
import pandas as pd
df = pd.read_csv('merged_DateFrames.csv',index_col=None,low_memory=False)
```

```python
# Convert 'NoticeSentBy' to datetime
df['NoticeSentBy'] = pd.to_datetime(df['NoticeSentBy'], format='%d/%m/%Y %H:%M')

# Group by 'GreyhoundName' and get the last row for each group
latest_updates = df.groupby('GreyhoundName').apply(lambda group: group.nlargest(1, 'NoticeSentBy'))


```

```python
latest_updates
```

```python
filtered_data = latest_updates[latest_updates['Retirement Type'] != 'RAC']
```

```python
filtered_data = filtered_data[filtered_data['Retirement Type'] != 'NEW']
```
