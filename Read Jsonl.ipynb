```python
import json

# Open the JSONL file for reading
with open('twitter.jsonl', 'r', encoding='utf-8') as file:
    json_objects = []  # Initialize an empty list to store parsed JSON objects
    
    # Iterate over lines in the file
    for line in file:
        # Parse each line as a JSON object
        try:
            data = json.loads(line)
            
            # Append the parsed object to the list
            json_objects.append(data)
        except json.JSONDecodeError as e:
            # Handle JSON decoding errors for invalid lines if needed
            print(f"Error decoding JSON: {e}")
```

```python
import pandas as pd
df = pd.DataFrame(json_objects)
```

```python
# set the index of the dataframe
df.set_index("id", inplace = True)
```

```python
# drop the second id column
df.drop("_id", axis='columns')
```

```python
# saving the dataframeserss
df.to_csv('twitter.csv')
```
