```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

# Load your dataset (replace 'data.csv' with your file path)
data = pd.read_csv('AllGreyhounds.csv')

# Select columns with categorical data
categorical_columns = data['Retirement Type']  # Replace with your categorical column names

# Perform one-hot encoding for categorical columns
data_encoded = pd.get_dummies(data, columns=categorical_columns)

# Separate data with missing and non-missing values
data_train = data_encoded[data_encoded['Retirement Type'].notnull()]
data_predict = data_encoded[data_encoded['Retirement Type'].isnull()]

# Split the training data into features (X) and target (y)
X_train = data_train.drop(categorical_columns, axis=1)
y_train = data_train['Retirement Type']

# Create and train a K-Nearest Neighbors classifier
k = 3  # Choose the number of neighbors
classifier = KNeighborsClassifier(n_neighbors=k)
classifier.fit(X_train, y_train)

# Predict missing values
X_predict = data_predict.drop(categorical_columns, axis=1)
predicted_categories = classifier.predict(X_predict)

# Fill missing values in the original dataset
data.loc[data['Retirement Type'].isnull(), 'Retirement Type'] = predicted_categories

# Now 'category' should have missing values filled with predictions
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier

# Load your dataset (replace 'data.csv' with your file path)
data = pd.read_csv('AllGreyhounds.csv')

# Select columns with categorical data
categorical_column = 'Retirement Type'  # Replace with your categorical column name

# Replace NaN values with a placeholder value (e.g., 'UNKNOWN')
data[categorical_column].fillna('UNKNOWN', inplace=True)

# Perform one-hot encoding for categorical columns
data_encoded = pd.get_dummies(data, columns=[categorical_column])

# Separate data with missing and non-missing values
data_train = data_encoded[data_encoded[categorical_column].ne('UNKNOWN')]
data_predict = data_encoded[data_encoded[categorical_column].eq('UNKNOWN')]

# Split the training data into features (X) and target (y)
X_train = data_train.drop(categorical_columns, axis=1)
y_train = data_train[categorical_columns]

# Create and train a K-Nearest Neighbors classifier
k = 3  # Choose the number of neighbors
classifier = KNeighborsClassifier(n_neighbors=k)
classifier.fit(X_train, y_train)

# Predict missing values
X_predict = data_predict.drop(categorical_columns, axis=1)
predicted_categories = classifier.predict(X_predict)

# Fill missing values in the original dataset
data.loc[data[categorical_column].eq('UNKNOWN'), categorical_column] = predicted_categories

# Now 'category' should have missing values filled with predictions
```
