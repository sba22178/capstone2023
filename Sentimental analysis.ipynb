```python
# Search the 'Title' column for the word 'example'
word_to_search = 'greyhound'
filtered_df = df[df['content'].str.contains(word_to_search, case=False)]
```

```python
filtered_df
```

```python
from textblob import TextBlob

def analyze_sentiment1(text):
    try:
        analysis = TextBlob(text)
        if analysis.sentiment.polarity > 0:
            return 'Positive'
        elif analysis.sentiment.polarity == 0:
            return 'Neutral'
        else:
            return 'Negative'
    except Exception as e:
        print(f"Error analyzing sentiment: {str(e)}")
        return 'Error'
```

```python
# Apply sentiment analysis to your DataFrame
filtered_df['Sentiment'] = filtered_df['content'].apply(analyze_sentiment1)
```

Sentiment Lexicons: Lexicon-based approaches use sentiment lexicons or dictionaries containing words with associated sentiment scores. Each word in a text is assigned a sentiment score, and the overall sentiment of the text is calculated based on the scores. Examples of lexicons include the AFINN lexicon and the VADER lexicon

```python
from afinn import Afinn
```

```python
# Create an Afinn object
afinn = Afinn()
```

```python
# Function to calculate sentiment scores for each row in the "Text" column
def calculate_sentiment(text):
    return afinn.score(text)
```

```python
# Apply the sentiment analysis function to the "Text" column and store the results in a new column "Sentiment"
#filtered_df['Sentiment_lex'] = filtered_df['Content'].apply(calculate_sentiment)

# Apply the sentiment analysis function to the "Text" column and store the results in a new column "Sentiment"
filtered_df['Sentiment_lex'] = filtered_df['content'].apply(calculate_sentiment)
```

```python
import matplotlib.pyplot as plt
from textblob import TextBlob
from afinn import Afinn

# Sample list of texts
texts = filtered_df['content']

# Initialize sentiment analyzers
textblob_analyzer = TextBlob("")
afinn_analyzer = Afinn()

# Initialize lists to store sentiment scores
textblob_scores = []
afinn_scores = []

# Calculate sentiment scores for each text
for text in texts:
    textblob_analysis = TextBlob(text)
    textblob_score = textblob_analysis.sentiment.polarity
    textblob_scores.append(textblob_score)

    afinn_score = afinn_analyzer.score(text)
    afinn_scores.append(afinn_score)

# Create a plot to compare sentiment scores
plt.figure(figsize=(10, 5))
plt.bar(range(len(texts)), textblob_scores, width=0.4, label='TextBlob (NLP)')
plt.bar([x + 0.4 for x in range(len(texts))], afinn_scores, width=0.4, label='AFINN (Lexicon)')
plt.xlabel('Texts')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis Comparison: TextBlob vs. AFINN')
plt.legend()
plt.show()
```
